```python
# =====================================================
# AGENTIC AI EV TARIFF OPTIMIZATION PIPELINE
# =====================================================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from xgboost import XGBRegressor

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# =====================================================
# FEATURE ENGINEERING
# =====================================================

print("Creating Time Features...")

urban_df['hour'] = urban_df['timestamp'].dt.hour
urban_df['day'] = urban_df['timestamp'].dt.day
urban_df['month'] = urban_df['timestamp'].dt.month
urban_df['weekday'] = urban_df['timestamp'].dt.weekday

urban_df['is_weekend'] = (
    urban_df['weekday'].isin([5, 6]).astype(int)
)

urban_df['peak_hour'] = urban_df['hour'].apply(
    lambda x: 1 if 17 <= x <= 22 else 0
)

urban_df['utilization_rate'] = (
    urban_df['occupancy'] /
    urban_df['occupancy'].max()
)

urban_df['congestion_score'] = (
    urban_df['occupancy'] *
    urban_df['volume']
)

urban_df['demand_intensity'] = (
    urban_df['volume'] /
    (urban_df['duration'] + 1)
)

urban_df = urban_df.sort_values(
    ['station_id', 'timestamp']
)

# =====================================================
# LAG FEATURES
# =====================================================

print("Creating Lag Features...")

urban_df['occupancy_lag_1'] = (
    urban_df.groupby('station_id')['occupancy']
    .shift(1)
)

urban_df['occupancy_lag_2'] = (
    urban_df.groupby('station_id')['occupancy']
    .shift(2)
)

urban_df['occupancy_lag_3'] = (
    urban_df.groupby('station_id')['occupancy']
    .shift(3)
)

urban_df['rolling_occupancy_mean'] = (
    urban_df.groupby('station_id')['occupancy']
    .transform(
        lambda x: x.rolling(3).mean()
    )
)

# =====================================================
# TARGET CREATION
# 3 HOUR FORECAST
# =====================================================

print("Creating 3-Hour Ahead Target...")

urban_df['future_occupancy'] = (
    urban_df.groupby('station_id')['occupancy']
    .shift(-36)
)

urban_df = urban_df.dropna()

# =====================================================
# FEATURE SET
# =====================================================

feature_cols = [
    'volume',
    'price',
    'duration',
    'hour',
    'weekday',
    'is_weekend',
    'peak_hour',
    'utilization_rate',
    'congestion_score',
    'demand_intensity',
    'occupancy_lag_1',
    'occupancy_lag_2',
    'occupancy_lag_3',
    'rolling_occupancy_mean'
]

target_col = 'future_occupancy'

X = urban_df[feature_cols]
y = urban_df[target_col]

print("\nFeature Shape:", X.shape)
print("Target Shape:", y.shape)

# =====================================================
# TIME SERIES SPLIT
# =====================================================

split_index = int(len(X) * 0.8)

X_train = X.iloc[:split_index]
X_test = X.iloc[split_index:]

y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]

print("\nTrain Shape:", X_train.shape)
print("Test Shape :", X_test.shape)

# =====================================================
# DEMAND PREDICTION AGENT
# =====================================================

print("\nTraining Demand Prediction Agent...")

model = XGBRegressor(
    n_estimators=150,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)

print("Model Training Completed")

# =====================================================
# PREDICTIONS
# =====================================================

y_pred = model.predict(X_test)

# =====================================================
# EVALUATION
# =====================================================

mae = mean_absolute_error(y_test, y_pred)

rmse = np.sqrt(
    mean_squared_error(y_test, y_pred)
)

r2 = r2_score(y_test, y_pred)

print("\n==============================")
print("DEMAND PREDICTION RESULTS")
print("==============================")

print("MAE :", round(mae, 4))
print("RMSE:", round(rmse, 4))
print("R²  :", round(r2, 4))

# =====================================================
# FEATURE IMPORTANCE
# =====================================================

importance_df = pd.DataFrame({
    'Feature': feature_cols,
    'Importance': model.feature_importances_
})

importance_df = (
    importance_df
    .sort_values(
        'Importance',
        ascending=False
    )
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=importance_df,
    x='Importance',
    y='Feature'
)

plt.title(
    "Demand Prediction Agent - Feature Importance"
)

plt.show()

# =====================================================
# ACTUAL VS PREDICTED
# =====================================================

sample_size = 500

plt.figure(figsize=(14,6))

plt.plot(
    y_test.values[:sample_size],
    label='Actual'
)

plt.plot(
    y_pred[:sample_size],
    label='Predicted'
)

plt.legend()

plt.title(
    "Actual vs Predicted Occupancy"
)

plt.show()

# =====================================================
# SAVE PREDICTIONS
# =====================================================

urban_df.loc[
    X_test.index,
    'predicted_occupancy'
] = y_pred

# =====================================================
# TARIFF PRICING AGENT
# =====================================================

print("\nRunning Dynamic Pricing Agent...")

BASE_PRICE = 15

urban_df['predicted_utilization'] = (
    urban_df['predicted_occupancy'] /
    urban_df['predicted_occupancy'].max()
) * 100


def dynamic_tariff(util):

    if util < 30:
        return 10

    elif util < 80:
        return 15

    else:
        return 20


urban_df['dynamic_tariff'] = (
    urban_df['predicted_utilization']
    .apply(dynamic_tariff)
)

# =====================================================
# REVENUE CALCULATION
# =====================================================

urban_df['fixed_revenue'] = (
    urban_df['volume'] *
    BASE_PRICE
)

urban_df['dynamic_revenue'] = (
    urban_df['volume'] *
    urban_df['dynamic_tariff']
)

old_revenue = (
    urban_df['fixed_revenue']
    .sum()
)

new_revenue = (
    urban_df['dynamic_revenue']
    .sum()
)

revenue_gain = (
    (new_revenue - old_revenue)
    / old_revenue
) * 100

# =====================================================
# UTILIZATION BUCKETS
# =====================================================

urban_df['utilization_bucket'] = pd.cut(
    urban_df['predicted_utilization'],
    bins=[0,30,80,100],
    labels=[
        'Low',
        'Medium',
        'High'
    ]
)

# =====================================================
# MONITORING AGENT
# =====================================================

print("\n==============================")
print("MONITORING AGENT")
print("==============================")

monitoring_results = {
    "Revenue Gain %":
        revenue_gain,

    "Average Utilization":
        urban_df['predicted_utilization']
        .mean(),

    "Peak Utilization":
        urban_df['predicted_utilization']
        .max(),

    "Average Dynamic Tariff":
        urban_df['dynamic_tariff']
        .mean()
}

for k,v in monitoring_results.items():
    print(f"{k}: {round(v,2)}")

# =====================================================
# VISUALIZATION
# =====================================================

plt.figure(figsize=(8,5))

urban_df[
    'utilization_bucket'
].value_counts().plot(
    kind='bar'
)

plt.title(
    "Predicted Utilization Categories"
)

plt.show()

# =====================================================
# OUTPUT FILES
# =====================================================

print("\nSaving Output Files...")

urban_df[
    [
        'timestamp',
        'station_id',
        'predicted_occupancy'
    ]
].to_csv(
    "demand_predictions.csv",
    index=False
)

urban_df[
    [
        'timestamp',
        'station_id',
        'dynamic_tariff'
    ]
].to_csv(
    "dynamic_tariffs.csv",
    index=False
)

monitor_df = pd.DataFrame(
    monitoring_results.items(),
    columns=[
        'Metric',
        'Value'
    ]
)

monitor_df.to_csv(
    "monitoring_results.csv",
    index=False
)

print("\nAll Deliverables Generated Successfully")
print("1. demand_predictions.csv")
print("2. dynamic_tariffs.csv")
print("3. monitoring_results.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# --- Phase 1: Preprocessing ---
def preprocess_urban_data():
    print("--- Phase 1: Preprocessing UrbanEV Data ---")
    df_occ = pd.read_csv(r'Datasets OP_26 Analytics\UrbanEV_ SZ_districts\occupancy.csv')
    df_time = pd.read_csv(r'Datasets OP_26 Analytics\UrbanEV_ SZ_districts\time.csv')
    df_stations = pd.read_csv(r'Datasets OP_26 Analytics\UrbanEV_ SZ_districts\stations.csv')

    # Align datetime
    df_time['datetime'] = pd.to_datetime(df_time[['year', 'month', 'day', 'hour', 'minute', 'second']])
    df_occ['datetime'] = df_time['datetime']
    df_occ = df_occ.drop(columns=['timestamp'])

    # Melt to long format
    df_occ_long = df_occ.melt(id_vars=['datetime'], var_name='station_id', value_name='occupancy_count')
    df_occ_long['station_id'] = pd.to_numeric(df_occ_long['station_id'])

    # Merge with station data
    df_urban = pd.merge(df_occ_long, df_stations, on='station_id', how='left')
    
    # Prevent Division by Zero
    df_urban['count'] = pd.to_numeric(df_urban['count'], errors='coerce')
    df_urban = df_urban[df_urban['count'] > 0].copy() 
    
    # Calculate Utilization Feature and cap at 1.0 (100%) to handle dirty data anomalies
    df_urban['utilization_rate'] = df_urban['occupancy_count'] / df_urban['count']
    df_urban['utilization_rate'] = df_urban['utilization_rate'].clip(upper=1.0)
    
    # Clean any mathematically invalid rows (infinities) before modeling
    df_urban.replace([np.inf, -np.inf], np.nan, inplace=True)
    df_urban = df_urban.dropna(subset=['utilization_rate'])
    
    return df_urban

def preprocess_acn_data():
    print("--- Phase 1: Preprocessing ACN Data ---")
    df_acn = pd.read_excel(r'Datasets OP_26 Analytics\ACN Data_ 25 April 2018 to 16 Dec 2018\acndata_sessions.json.xlsx')
    
    # Calculate Session Durations and baseline revenue
    df_acn['connectionTime'] = pd.to_datetime(df_acn['connectionTime'])
    df_acn['disconnectTime'] = pd.to_datetime(df_acn['disconnectTime'])
    df_acn['session_duration_hrs'] = (df_acn['disconnectTime'] - df_acn['connectionTime']).dt.total_seconds() / 3600
    df_acn['baseline_revenue_inr'] = df_acn['kWhDelivered'] * 15 # Flat rate baseline
    
    # Clean failed sessions
    df_acn_clean = df_acn[df_acn['kWhDelivered'] > 0].copy()
    return df_acn_clean

# --- Phase 1.5: Feature Engineering ---
def engineer_features(df_urban):
    print("--- Phase 1.5: Engineering Robust Temporal Features ---")
    # Ensure sequential ordering for time-series features
    df_urban = df_urban.sort_values(by=['station_id', 'datetime'])
    
    df_urban['hour'] = df_urban['datetime'].dt.hour
    df_urban['day_of_week'] = df_urban['datetime'].dt.dayofweek
    df_urban['is_weekend'] = df_urban['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
    
    # 1-hour temporal lag (12 intervals of 5 minutes = 1 hour)
    df_urban['lag_1h_utilization'] = df_urban.groupby('station_id')['utilization_rate'].shift(12)
    
    # 3-hour rolling average to capture the broader trend and smooth noise
    df_urban['rolling_3h_utilization'] = df_urban.groupby('station_id')['utilization_rate'].transform(lambda x: x.rolling(window=36, min_periods=1).mean())
    
    return df_urban

# --- Phase 2: Exploratory Data Analysis (EDA) ---
def run_eda(df_urban, df_acn_clean):
    print("\n--- Phase 2: Running EDA ---")
    
    hourly_utilization = df_urban.groupby('hour')['utilization_rate'].mean().reset_index()
    
    plt.figure(figsize=(10, 5))
    sns.lineplot(data=hourly_utilization, x='hour', y='utilization_rate', marker='o')
    plt.title('Average Intraday Charger Utilization Rate (Shenzhen)')
    plt.xlabel('Hour of Day')
    plt.ylabel('Utilization Rate')
    plt.grid(True)
    plt.savefig('intraday_utilization.png')
    plt.close()
    print("Saved intraday_utilization.png")

    plt.figure(figsize=(10, 5))
    sns.histplot(df_acn_clean['session_duration_hrs'], bins=50, kde=True)
    plt.title('Distribution of Charging Session Durations')
    plt.xlabel('Session Duration (Hours)')
    plt.xlim(0, 24)
    plt.ylabel('Frequency')
    plt.savefig('session_durations.png')
    plt.close()
    print("Saved session_durations.png")

# --- Phase 3 & 4: Robust Demand Prediction Agent (Model Training) ---
def train_demand_model(df_urban):
    print("\n--- Phase 3 & 4: Training Robust Demand Prediction Model ---")
    
    # Drop rows where lag features couldn't be calculated (first hour of data per station)
    model_df = df_urban.dropna(subset=['lag_1h_utilization', 'utilization_rate']).copy()
    
    # Notice: 'station_id' is removed to force generalization rather than memorization
    features = ['hour', 'day_of_week', 'is_weekend', 'lag_1h_utilization', 'rolling_3h_utilization']
    
    # Chronological Train-Test Split (Prevents Data Leakage/Overfitting)
    model_df = model_df.sort_values(by='datetime')
    split_idx = int(len(model_df) * 0.8)
    
    train_data = model_df.iloc[:split_idx]
    test_data = model_df.iloc[split_idx:]
    
    X_train = train_data[features]
    y_train = train_data['utilization_rate']
    X_test = test_data[features]
    y_test = test_data['utilization_rate']
    
    # Regularized Random Forest
    model = RandomForestRegressor(
        n_estimators=100,          
        max_depth=8,              # Reduced depth
        min_samples_split=20,     # Increased split requirement
        min_samples_leaf=10,      # Increased leaf requirement
        random_state=42, 
        n_jobs=-1
    )
    model.fit(X_train, y_train)
    
    predictions = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    mae = mean_absolute_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    
    print("Robust Model Evaluation Metrics (Generalization Ensured):")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE:  {mae:.4f}")
    print(f"  R^2:  {r2:.4f}")
    
    return model

# --- Phase 5: Dynamic Pricing & Monitoring Simulation ---
def simulate_network(df_urban, trained_model):
    print("\n--- Phase 5: Running Agentic Feedback Simulation ---")
    
    # Validate on chronological trailing window
    val_data = df_urban.dropna(subset=['lag_1h_utilization', 'utilization_rate']).sort_values(by='datetime').tail(2000).copy()
    
    simulation_results = []
    baseline_revenue_total = 0
    agent_revenue_total = 0
    off_peak_discount_triggers = 0
    
    for _, row in val_data.iterrows():
        # Match the exact features the robust model was trained on
        features = [[row['hour'], row['day_of_week'], row['is_weekend'], row['lag_1h_utilization'], row['rolling_3h_utilization']]]
        predicted_util = trained_model.predict(features)[0]
        
        base_rate = 15.0
        if predicted_util > 0.8:
            tariff = base_rate * 1.5 
        elif predicted_util < 0.3:
            tariff = base_rate * 0.7 
            off_peak_discount_triggers += 1
        else:
            tariff = base_rate
            
        assumed_kwh = 30 
        baseline_rev = base_rate * assumed_kwh
        agent_rev = tariff * assumed_kwh
        
        baseline_revenue_total += baseline_rev
        agent_revenue_total += agent_rev
        
        simulation_results.append({
            'timestamp': row['datetime'],
            'station_id': row['station_id'],
            'actual_utilization': row['utilization_rate'],
            'predicted_utilization': predicted_util,
            'applied_tariff': tariff,
            'baseline_revenue': baseline_rev,
            'agent_revenue': agent_rev
        })
        
    results_df = pd.DataFrame(simulation_results)
    results_df.to_csv('simulation_results.csv', index=False)
    print("Saved simulation_results.csv")
    
    revenue_gain_pct = ((agent_revenue_total - baseline_revenue_total) / baseline_revenue_total) * 100
    
    print("\n--- Final Business Metrics ---")
    print(f"  Total Baseline Revenue: ₹{baseline_revenue_total:.2f}")
    print(f"  Total Agent Revenue:    ₹{agent_revenue_total:.2f}")
    print(f"  Revenue Gain %:         {revenue_gain_pct:.2f}%")
    print(f"  Off-Peak Uplift Triggers: {off_peak_discount_triggers} sessions")

# --- CORE EXECUTION FLOW ---
if __name__ == "__main__":
    df_urban_processed = preprocess_urban_data()
    df_acn_processed = preprocess_acn_data()
    
    # Ensure features are applied globally to df_urban_processed
    df_urban_featured = engineer_features(df_urban_processed)
    
    run_eda(df_urban_featured, df_acn_processed)
    
    trained_model = train_demand_model(df_urban_featured)
    
    simulate_network(df_urban_featured, trained_model)
```
